```python
aim="prints whether the number is palindrome or not palindrome"
print(aim)
number = input("Enter a number: ")
if number == number[::-1]:
 print("Palindrome")
else:
 print("Not Palindrome")
for i in range(10):
 count = number.count(str(i))
if count > 0:
 print(f"{i} appears {count} times")
```
